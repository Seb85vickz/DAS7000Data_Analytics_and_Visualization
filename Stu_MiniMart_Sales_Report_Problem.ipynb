# **Case Study: MiniMart Sales Report Problem**

Business Context:
MiniMart is a small retail store that sells everyday items. The manager wants a simple Python program to analyze daily sales using data structures and basic libraries.

The store has the following data:

Product Catalog (List of Tuples)

Each product has: product_id, name, price.
Example:

```python
products = [
    (101, "Soap", 40),
    (102, "Shampoo", 120),
    (103, "Toothpaste", 60),
    (104, "Hair Oil", 90)
]
```

Sales Data (Dictionary)

Key → product_id

Value → quantity sold today.
Example:

```python
sales_today = {
    101: 15,
    102: 5,
    103: 10
}
```

Manager Needs the Report to Show:

(a) Total Revenue (quantity × price for each product).

(b) Unique Products Sold Today (use a set).

(c) Best-Selling Product (highest quantity).

(d) Products Sorted by Price.

(e) A Pandas DataFrame showing sales with revenue column.  In a single dataframe show - Product_ID, Name,  Price,  Quantity,  Revenue. Use NumPy for revenue calculation.

**Your output should look like this-**

sc1.png

```python
import numpy as np
import pandas as pd

# Product catalog: (product_id, name, price)
products = [
    (101, "Soap", 40),
    (102, "Shampoo", 120),
    (103, "Toothpaste", 60),
    (104, "Hair Oil", 90)
]

# Sales data: product_id → quantity sold today
sales_today = {
    101: 15,
    102: 5,
    104: 10
}

# Step 1: Create a list of sales records with revenue
sales_records = []
for pid, name, price in products:
    quantity = sales_today.get(pid, 0)
    revenue = price * quantity
    sales_records.append((pid, name, price, quantity, revenue))

# Step 2: Create DataFrame
df = pd.DataFrame(sales_records, columns=["Product_ID", "Name", "Price", "Quantity", "Revenue"])

# Step 3: Total Revenue
total_revenue = df["Revenue"].sum()

# Step 4: Unique Products Sold
unique_products = list(set(df[df["Quantity"] > 0]["Name"]))

# Step 5: Best-Selling Product
best_seller_row = df.loc[df["Quantity"].idxmax()]
best_seller = best_seller_row["Name"]

# Step 6: Products Sold (Name, Quantity)
products_sold = [(row["Name"], row["Quantity"]) for _, row in df.iterrows() if row["Quantity"] > 0]

# Final Report
print("MiniMart Daily Sales Report")
print(f"Total Revenue: {total_revenue}")
print(f"Unique Products Sold: {unique_products}")
print(f"Best-Selling Product: '{best_seller}'")
print(f"Products Sold: {products_sold}")
print("\nSales DataFrame with Revenue:")
print(df)

```
