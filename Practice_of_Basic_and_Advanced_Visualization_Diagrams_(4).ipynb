```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

plt.style.use('ggplot')#Sets a light gray background with white gridlines. It makes plots easier to read compared to the default plain white background.
```

Basic Line Plots â€” Understanding Functional Relationships

```python
x = np.linspace(0, 20, 400)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x, y1, label='sin(x)', color='teal', linewidth=2)
ax.plot(x, y2, label='cos(x)', color='coral', linestyle='--', linewidth=2)
ax.set_title('Trigonometric Functions: Sine & Cosine', fontsize=14)
ax.set_xlabel('X Values')
ax.set_ylabel('Function Output')
ax.legend()
plt.show()
```

This plot compares two periodic functions  sine and cosine over the same range.

Observe amplitude symmetry, learn about multiple line overlays and legends.

**Subplots - Display Multiple Related Charts**

A 2Ã—2 grid compares different mathematical functions.

```python
x = np.linspace(0, 10, 200)

fig, axs = plt.subplots(2, 2, figsize=(10,6))
axs[0, 0].plot(x, np.sin(x), color='darkgreen')
axs[0, 0].set_title('sin(x)')

axs[0, 1].plot(x, np.cos(x), color='royalblue')
axs[0, 1].set_title('cos(x)')

axs[1, 0].plot(x, np.tan(x), color='crimson')
axs[1, 0].set_title('tan(x)')

axs[1, 1].plot(x, np.exp(-x), color='purple')
axs[1, 1].set_title('exp(-x)')

plt.suptitle('Common Mathematical Functions', fontsize=15)
plt.tight_layout()
plt.show()
```

**About tight layout**

When you create a figure with multiple subplots or long labels, elements may overlap or get clipped.

plt.tight_layout() calculates the optimal spacing and adjusts the subplot parameters (left, right, top, bottom, wspace, hspace) automatically.

It does not change the figure content, only the spacing around it.

**Grouped and Stacked Bars - Comparative Analysis**

Observe comparative patterns between two categories (Men vs. Women) across regions.

It introduces grouped bar charts - great for side-by-side comparisons.

```python

regions = ['North', 'South', 'East', 'West']
men_sales = [220, 270, 300, 250]
women_sales = [200, 240, 260, 280]

x = np.arange(len(regions))
width = 0.35

fig, ax = plt.subplots(figsize=(8,5))
ax.bar(x - width/2, men_sales, width, label='Men', color='lightblue')## Moves Men's bars slightly left
ax.bar(x + width/2, women_sales, width, label='Women', color='salmon') ## Moves Women's bars slightly right
ax.set_xticks(x)
ax.set_xticklabels(regions)
ax.set_ylabel('Units Sold')
ax.set_title('Sales by Gender & Region')
ax.legend()
plt.show()
```

Now students should learn, how to plot group bar chart with three or four groups

**Scatter Plot â€” Relationship Between Two Continuous Variables**

Demonstrates linear correlation - as x increases, y tends to increase.

Learn to spot positive/negative correlation visually.

```python
np.random.seed(10)
x = np.random.randn(1000)
y = 0.5*x + np.random.randn(1000)*0.3

fig, ax = plt.subplots(figsize=(6,5))
ax.scatter(x, y, alpha=0.8, c='steelblue', edgecolor='black')
ax.set_title('Scatter Plot: Linear Relationship')
ax.set_xlabel('X Variable')
ax.set_ylabel('Y Variable')
plt.show()
```

**Boxplot & Violin Plot - Distribution & Outliers**

These plots visualize spread, median, and outliers.

Violin plots add density information

Violin plots = Combines boxplot + kernel density estimation.

The width of the violin at each y-value shows density of points (more data â†’ wider shape).

```python
data = [np.random.normal(60, 10, 200), #np.random.normal(mean, std, n) generates n random numbers from a normal distribution:
        np.random.normal(75, 15, 200),
        np.random.normal(90, 20, 200)]

fig, axs = plt.subplots(1, 2, figsize=(12,5))

axs[0].boxplot(data)
axs[0].set_title('Boxplot of Distributions')
axs[0].set_xticklabels(['A', 'B', 'C'])

sns.violinplot(data=data, ax=axs[1], palette='muted')#colors the violins with a muted palette for clarity.
axs[1].set_title('Violin Plot of Distributions')
axs[1].set_xticklabels(['A', 'B', 'C'])

plt.show()
```

**Compare Violin A, B, C:**

* A is concentrated around 60 (narrow violin).
* B is wider (more spread).
* C is very wide and skewed (high std 20).

**Bubble Chart - Combine Three Variables**

Bubble size conveys a third dimension (pollution level).

Learn to represent multi-dimensional data visually.

```python
np.random.seed(1)
cities = ['Delhi', 'Mumbai', 'Kolkata', 'Bengaluru', 'Chennai', 'Pune']
population = np.random.randint(5, 20, len(cities)) * 1e6
gdp = np.random.rand(len(cities)) * 200 #np.random.rand(len(cities))- generates random floats between 0 and 1 for each city.
pollution = np.random.randint(50, 200, len(cities)) #Generates random integers between 50 and 199 to represent pollution levels.

plt.figure(figsize=(8,6))
plt.scatter(gdp, population, s=pollution, alpha=0.6, color='tomato', edgecolor='black')
plt.title('City GDP vs Population (Bubble Size = Pollution Index)')
plt.xlabel('GDP (Billion USD)')
plt.ylabel('Population')
for i, city in enumerate(cities):
    plt.text(gdp[i]+1, population[i], city)#gdp[i]+1 : shifts the text slightly right so it doesnâ€™t overlap the bubble. population[i] : vertical position of the text.
plt.show()
```

#Heatmap with Pivot Table (Sales by Product and Month)

* Heatmaps visualize data in a grid format, with color intensity representing values.
* Often created from a pivot table to compare two dimensions, e.g., products vs months.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data
dates = pd.date_range('2025-01-01', periods=12, freq='M')
products = ['Coffee', 'Tea', 'Bakery']
sales = np.random.randint(50, 150, (12,3))#12Ã—3 array of random integers between 50 and 149 these represent monthly sales for each product.

df_area = pd.DataFrame(sales, columns=products, index=dates)#Rows : months (dates), Columns : products (Coffee, Tea, Bakery), Each cell : monthly sales of that product


```

```python
import seaborn as sns

# Sample pivot
df_pivot = df_area.T  # transpose for heatmap (products as rows)
sns.heatmap(df_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Sales by Product')
plt.xlabel('Month')
plt.ylabel('Product')
plt.show()
```

**Interpretation:**

* Quickly spot high or low values.
* Darker colors = higher sales, lighter = lower sales.
* Highlights seasonal trends or underperforming products.

#3D Surface Plot
* Plots three variables in 3D space: X, Y, and Z.
* The surface represents the relationship between two independent variables and one dependent variable.

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))#Thatâ€™s a radially symmetric wave pattern, meaning the height depends only on the distance from the origin (0,0), not direction.
#As you move away from the center, Z oscillates like circular ripples spreading outward. So Z forms concentric wave-like hills and valleys.

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('3D Surface Plot Example')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

#Sunburst Chart

* A radial, hierarchical version of a treemap.
* Inner rings = top-level categories, outer rings = subcategories.


```python
import plotly.express as px

df = pd.DataFrame({
    'Category': ['Coffee','Coffee','Tea','Tea','Bakery','Bakery'],
    'Subcategory': ['Latte','Espresso','Green Tea','Black Tea','Croissant','Muffin'],
    'Sales': [500, 700, 300, 500, 400, 550]
})

fig = px.sunburst(df, path=['Category','Subcategory'], values='Sales', color='Sales', color_continuous_scale='Viridis')
fig.show()
```

Here for Bakery, Coffee and Tea, numbers are rescaled to fit the circular area mathematically.

**Interpretation:**

#Stacked Area Chart

* A stacked area chart shows how different categories contribute to a total over time.
* Each layer represents a category, stacked on top of each other.
* The height of the top line represents the total value at that point in time.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data
dates = pd.date_range('2025-01-01', periods=12, freq='M')
products = ['Coffee', 'Tea', 'Bakery']
sales = np.random.randint(50, 150, (12,3))#12Ã—3 array of random integers between 50 and 149 these represent monthly sales for each product.

df_area = pd.DataFrame(sales, columns=products, index=dates)#Rows : months (dates), Columns : products (Coffee, Tea, Bakery), Each cell : monthly sales of that product

# Plot
df_area.plot(kind='area', stacked=True, alpha=0.7)
plt.title('Monthly Cumulative Sales by Product')
plt.ylabel('Sales ($)')
plt.xlabel('Month')
plt.show()
```

**Interpretation:**

* Identify which products drive overall sales trends.
* Peaks indicate high sales periods.
* Useful for visualizing cumulative contributions over months or weeks.

#Streamgraph (Stacked Area with Emphasis on Flow)
* A variation of stacked area charts that emphasize flow over time.
* Instead of starting from zero, layers are centered around an axis, these create a flowing effect.

```python
import matplotlib.cm as cm #colormap: a module that provides predefined color palettes (like tab10, viridis, plasma, etc.).

# Sample data
categories = ['Coffee','Tea','Bakery']
sales_data = np.random.randint(20, 100, (12,3)) #generates a 12Ã—3 matrix of random integers between 20 and 100.

fig, ax = plt.subplots(figsize=(12,6))
colors = cm.tab10(np.arange(len(categories)))

ax.stackplot(range(12), sales_data.T, labels=categories, colors=colors, alpha=0.8)
#sales_data.T â†’ Transposes the matrix, that converts shape from (12,3) â†’ (3,12).
#Why? Because stackplot() expects each row to be a separate categoryâ€™s y-values.
ax.legend(loc='upper left')
plt.title('Streamgraph of Monthly Sales')
plt.show()
```

Interpretation:
* Visualizes relative changes over time and identifies trends.
* Useful for understanding dominance shifts among categories.

#Parallel Coordinates Plot (Multivariate Analysis)
* Multivariate visualization.
* Each variable has its own vertical axis.
* Lines connecting the axes represent individual observations.

```python
from pandas.plotting import parallel_coordinates

# Sample multivariate dataset
df_parallel = pd.DataFrame({
    'Coffee':[120,130,150,110],
    'Tea':[80,85,90,70],
    'Bakery':[60,75,80,65],
    'Store':['A','B','C','D']
})

parallel_coordinates(df_parallel, 'Store', colormap='Set2')
plt.title('Parallel Coordinates: Sales Comparison Across Stores')
plt.show()
```

**Interpretation:**

* Compare patterns across multiple variables simultaneously.
* Identify correlations or anomalies.
* For example, a store with high coffee sales but low bakery sales stands out as a line will diverge from others.

#Interactive Scatter Plot with Hover & Color Encoding

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Sample dataset
np.random.seed(10)
df = pd.DataFrame({
    'Revenue': np.random.randint(50, 200, 50),
    'Profit': np.random.randint(10, 80, 50),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 50),
    'Store': [f'Store_{i}' for i in range(1, 51)]
})

# Interactive scatter plot
fig = px.scatter(
    df,
    x='Revenue',
    y='Profit',
    color='Region',
    hover_data=['Store'],#This explicitly tells Plotly: Include the Store column in the hover tooltip.
    size='Profit',
    title='Interactive Revenue vs Profit by Region',
    color_discrete_sequence=px.colors.qualitative.Pastel#it is one of Plotlyâ€™s predefined qualitative color sets, which are soft pastel colors.
)

fig.show()
```

#Interactive Time-Series Line Chart with Range Slider

```python
import plotly.graph_objects as go #Used to create fully customizable interactive plots.
import pandas as pd
import numpy as np

# Sample time series data
dates = pd.date_range('2024-01-01', periods=200)
sales = np.random.randint(200, 500, size=200)

fig = go.Figure() #Initializes a blank Plotly figure to which we can add traces (lines, bars, markers, etc.).
fig.add_trace(go.Scatter( #Adds a scatter plot, which can be connected with lines.
    x=dates, y=sales, mode='lines+markers', #Plots both lines connecting points and markers at each point.
    line=dict(color='royalblue'),
    name='Daily Sales'
))

fig.update_layout(
    title='Interactive Time Series of Daily Sales',
    xaxis_title='Date',
    yaxis_title='Sales ($)',
    xaxis=dict(rangeslider=dict(visible=True)),
    template='plotly_white' #Uses a clean white background theme.
)

fig.show()
```

**What the chart shows**

* A time series of daily sales over 200 days.
* Each point is a day, connected by a royal blue line.
* Hovering over points shows exact date and sales value.
* The range slider allows for selecting specific time intervals for detailed analysis.

#Interactive Treemap (Hierarchical Sales Representation)

```python
import plotly.express as px
import pandas as pd

# Sample hierarchical data
df = pd.DataFrame({
    'Category': ['Coffee','Coffee','Tea','Tea','Bakery','Bakery'],
    'Subcategory': ['Latte','Espresso','Green Tea','Black Tea','Croissant','Muffin'],
    'Sales': [500, 700, 300, 500, 400, 550]
})

# Interactive Treemap
fig = px.treemap(
    df,
    path=['Category', 'Subcategory'],
    values='Sales',
    color='Sales',
    color_continuous_scale='viridis',
    title='Interactive Treemap: Sales by Category and Subcategory'
)
fig.show()
```

**Interpretation**

* Each rectangleâ€™s size = sales volume.
* Each color shade = relative performance (darker = higher).
* Click to drill down from category to subcategory.
* Excellent for hierarchical sales breakdowns (department to product).

#Interactive 3D Scatter Plot

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Create synthetic 3D data
np.random.seed(42)
df = pd.DataFrame({
    'Revenue': np.random.randint(100, 500, 50),
    'Profit': np.random.randint(20, 120, 50),
    'Expenses': np.random.randint(50, 300, 50),
    'Region': np.random.choice(['North','South','East','West'], 50)
})

fig = px.scatter_3d(
    df,
    x='Revenue', y='Profit', z='Expenses',
    color='Region',
    size='Profit',
    hover_data=['Region'],
    title='3D Interactive Scatter: Business Performance Overview'
)

fig.show()
```

**Interpretation**

* Each point is a business unit or branch.
* You can rotate and zoom to view patterns in 3D space.
* Quickly visualize how Revenue, Profit, and Expenses interact.
* Regions color-coded for easier comparison.
